```python
import numpy as np

a = np.array([[0.,0.,0.],
              [10.,10.,10.],
              [20.,20.,20.],
              [30.,30.,30.]])

b = np.array([0., 1., 2.])

c = a + b
print("a + b:")
print(c)

mask = (c >= 22)
print("Mask (c >= 22):")
print( mask)

count = np.sum(mask)
print("Count of values >= 22:", count) 
```

```python
grades = np.array([[87, 96, 70],
                   [100, 87, 90],
                   [94, 77, 90],
                   [100, 81, 82]])
print("Sum:", np.sum(grades))
print("Min:", np.min(grades))
print("Max:", np.max(grades))
print("Mean:", np.mean(grades))
print("Std:", np.std(grades))
print("Var:", np.var(grades))

print("Mean per column:", np.mean(grades, axis=0))
print("Mean per row:", np.mean(grades, axis=1))
print("Max per row:", np.max(grades, axis=1))
```

```python
arr = np.array([[2, 4, 6],
                [1, 3, 5],
                [7, 9, 11]])

print("Array:\n", arr)
print("Sum of each row:", np.sum(arr, axis=1))
print("Sum of each column:", np.sum(arr, axis=0))
print("Overall mean:", np.mean(arr))
```

```python
A = np.array([[10, 20, 30],
              [40, 50, 60]])

B = np.array([1, 2, 3])

print("A + B:\n", A + B)
print("A * 2:\n", A * 2)
print("A - 5:\n", A - 5)
```

```python
grades = np.array([[70, 85, 90],
                   [88, 76, 95],
                   [60, 80, 72]])

print("\nIndex of largest value in whole array:",
      np.argmax(grades))

print("Index of highest grade in each row:",
      np.argmax(grades, axis=1))

print("Index of lowest grade in each column:",
      np.argmin(grades, axis=0))
```
